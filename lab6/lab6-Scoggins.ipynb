```python
def SIM(N, k=5):
    rr = np.random.rand(N) * k
    TF = rr > k/2.
    return TF
```

```python
test = SIM(50)
test
```

```python
plt.plot(test)
```

```python
P = np.cumsum(test, dtype='float') / np.arange(1,51)

plt.plot(P)
plt.xlabel('Number of trials')
plt.ylabel('P')
```
