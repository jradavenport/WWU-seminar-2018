```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.rand(10)
```

```python
random1 = np.random.rand(10000) * 20.

_ = plt.hist(random1)
plt.xlabel('Simulated Value')
plt.ylabel('Number of Trials')
```

```python
random2 = np.random.randn(10000) * 20.

_ = plt.hist(random2, bins=20)
plt.xlabel('Simulated Value')
plt.ylabel('Number of Trials')
```

```python
def SIM(N, k=5):
    rr = np.random.rand(N) * k
    TF = rr > k/2.
    return TF
```

```python
test = SIM(50)
test
```

```python
plt.plot(test)
```

```python
P = np.cumsum(test, dtype='float') / np.arange(1,51)

plt.plot(P)
plt.xlabel('Number of trials')
plt.ylabel('P')
```
