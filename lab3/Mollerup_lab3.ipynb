In general you'll try to [catch these](https://wiki.python.org/moin/HandlingExceptions) and handle them gracefully. Usually I write Try/Except statements as I find these errors


```python
def number_test(input):
    '''
    An example of using a Try/Except statement to catch an error.
    You could imagine doing something more useful than just printing stuff to the screen...
    '''
    try:
        val = float(input)
        print('Yup...')
    except ValueError:
        print("Ack! Not a float!")
        val = np.nan
        
    return val
```

```python
def timeTest(n):
    '''
    This function will take as input the the number of coordinates to test. It will generate a list of n random
    run them through the multiple different constellation search functions, and then output the time each function 
    took to locate each constellation
    
    Parameters
    ----------
    n: the number of random coordinates to generate and test
        
    Returns
    -------
    float : runtime of each method (Davenport, Scoggins respectively)
    '''
    decMax = 90
    decMin = -90
    declinations = np.random.random(n) * (decMax - decMin) + decMin
    raMax = 24
    raMin = 0
    rAscensions = np.random.random(n) * (raMax - raMin) + raMin
    
    dav_ti = time.time()
    solutions.constell_davenport(rAscensions, declinations)
    dav_tf = time.time()
    dav_runTime = dav_tf - dav_ti
    
    sco_ti = time.time()
    for x in range(len(declinations)):
        dec = declinations[x]
        ra = rAscensions[x]
        solutions.constell_scoggins(ra, dec)
    sco_tf = time.time()
    sco_runTime = sco_tf - sco_ti
    
    return  dav_runTime, sco_runTime
```

```python
dav, sco = timeTest(2000)
print(dav, sco)
```
